```python
import numpy as np
from sklearn.metrics import matthews_corrcoef
import matplotlib.pyplot as plt
```

```python
shape = 1000
num_iterations = 200
# np.random.seed(2023)
fig, ax = plt.subplots(figsize=(10,10))
plt.xlabel("MCC without averaging")
plt.ylabel("MCC with averaging")
plt.xlim(-0.1, 0.1)
plt.ylim(-0.1, 0.1)
points = []

for i in range(num_iterations):
	a = np.random.randint(2, size=(shape))
	b = np.random.randint(2, size=(shape))
	# x
	score1 =  matthews_corrcoef(a, b)

	a = a.reshape(100,10)
	b = b.reshape(100, 10)

	avging_scores = []
	for (_a, _b) in zip(a, b):
		avging_scores.append(matthews_corrcoef(_a, _b))
	mean_score = np.mean(avging_scores)

	points.append((score1,mean_score))

ax.plot([0,1],[0,1], transform=ax.transAxes, color='red')
x, y = zip(*points)
plt.scatter(x, y)
plt.show()
```

```python
prob = 0.05
np.random.choice([0, 1], size=shape, p=[1-prob, prob])

shape = 1000
num_iterations = 200
fig, ax = plt.subplots(figsize=(7,7))
plt.xlabel("MCC without averaging")
plt.ylabel("MCC with averaging")
plt.xlim(-0.1, 0.1)
plt.ylim(-0.1, 0.1)
points = []

for i in range(num_iterations):
	a = np.random.choice([0, 1], size=shape, p=[1-prob, prob])
	b = np.random.choice([0, 1], size=shape, p=[1-prob, prob])
	# x
	score1 =  matthews_corrcoef(a, b)

	a = a.reshape(100,10)
	b = b.reshape(100, 10)

	avging_scores = []
	for (_a, _b) in zip(a, b):
		avging_scores.append(matthews_corrcoef(_a, _b))
	mean_score = np.mean(avging_scores)

	points.append((score1,mean_score))

ax.plot([0,1],[0,1], transform=ax.transAxes, color='red')
x, y = zip(*points)
plt.scatter(x, y)
plt.show()

above_line = 0
for pair in points:
	if pair[0] < pair[1]:
		above_line += 1

print(f"Num of points above line: {above_line}/{num_iterations}")
```
